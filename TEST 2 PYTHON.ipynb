```python
#TASK 1
#ADDITIONAL : GIVE GEOMETRICAL INTERPRETATION OF THE SAME DATA
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the coefficients and right-hand side of the equations
coefficients = np.array([[1, 1, 1], [6, -4, 5], [5, 2, 2]])
rhs = np.array([2, 31, 13])

# Solve the system of equations to get the intersection point
solution = np.linalg.solve(coefficients, rhs)

# Extract the solutions for x, y, and z
x = solution[0]
y = solution[1]
z = solution[2]

# Create a figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define a grid of points for visualization
x_range = np.linspace(-5, 5, 10)
y_range = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x_range, y_range)

# Calculate the corresponding Z values for each plane
Z1 = 2 - X - Y
Z2 = (31 - 6*X + 4*Y) / 5
Z3 = (13 - 5*X - 2*Y) / 2

# Plot the planes
ax.plot_surface(X, Y, Z1, alpha=0.5, rstride=100, cstride=100, label='x + y + z = 2')
ax.plot_surface(X, Y, Z2, alpha=0.5, rstride=100, cstride=100, label='6x - 4y + 5z = 31')
ax.plot_surface(X, Y, Z3, alpha=0.5, facecolors='g', rstride=100, cstride=100, label='5x + 2y + 2z = 13')

# Plot the intersection point
ax.scatter(x, y, z, color='red', s=100, label='Intersection Point (3, -2, 1)')

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()

```

```python
#TASK 3
#3.1. Create a Dataframe 'emp_df' from the Dictionary 'emp_dict' [05 marks]
#3.2. Update the Dataframe 'emp_df' with the following Information: [05 marks]
#3.3. Update the Dataframe emp_df with the following information
#3.4. Create a Column Gross_Salary_Lakhs with the following information: [05 marks] 
#Gross_Salary_Lakhs = Salary_Lakhs * (1 + Bonus%)
import pandas as pd

# Create the Employee Dictionary with initial information
emp_dict = {
    'Name': ['ABC', 'DEF', 'GHI', 'JKL', 'MNO', 'PQR', 'STU'],
    'Education': ['Graduate', 'Postgraduate', 'Postgraduate', 'Graduate', 'Graduate', 'Postgraduate', 'Graduate'],
    'Gender': ['Male', 'Female', 'Male', 'Other', 'Female', 'Female', 'Male'],
    'Age': [22, 27, 26, 23, 24, 30, 21]
}

# Create a DataFrame from the dictionary
emp_df = pd.DataFrame(emp_dict)

# Update the DataFrame 'emp_df' with Salary_Lakhs and Bonus% information
salary_bonus = {
    'Salary_Lakhs': [6, 15, 20, 5, 10, 18, 12],
    'Bonus%': [12.50, 8.75, 6.25, 10.20, 13.60, 11.40, 9.80]
}
emp_df = emp_df.assign(**salary_bonus)

# Update the DataFrame 'emp_df' with additional employee information
new_employees = {
    'Name': ['VWX', 'YZA', 'BCD'],
    'Education': ['Postgraduate', 'Graduate', 'Postgraduate'],
    'Gender': ['Male', 'Female', 'Other'],
    'Age': [35, 28, 32],
    'Salary_Lakhs': [14, 7, 8],
    'Bonus%': [5.50, 7.75, 14.80]
}
emp_df = emp_df.append(pd.DataFrame(new_employees), ignore_index=True)

# Create a new column 'Gross_Salary_Lakhs' based on the given formula
emp_df['Gross_Salary_Lakhs'] = emp_df['Salary_Lakhs'] * (1 + emp_df['Bonus%'] / 100)

# Display the updated DataFrame 'emp_df'
print(emp_df)


```

```python
#TASK 3
#ADDITIONAL : TWO TABLES ONE HAVING SALARY MORE THAN 10 AND OTHER LESS THAN 10
# Create a DataFrame for employees with Salary_Lakhs >= 10
high_salary_df = emp_df[emp_df['Salary_Lakhs'] >= 10]

# Create a DataFrame for employees with Salary_Lakhs < 10
low_salary_df = emp_df[emp_df['Salary_Lakhs'] < 10]

# Display the two separate DataFrames
print("Employees with Salary_Lakhs >= 10:")
print(high_salary_df)

print("\nEmployees with Salary_Lakhs < 10:")
print(low_salary_df)

```

```python
#TASK 6
#ADDITIONAL : FIND THE AVERAGE AGE OF EMLOYEES WHO ARE GRADUATES AND POST GRADUATES
import pandas as pd

# Create the DataFrame with the given data
data = {
    'Name': ['ABC', 'DEF', 'GHI', 'JKL', 'MNO', 'PQR', 'STU'],
    'Education': ['Graduate', 'Postgraduate', 'Postgraduate', 'Graduate', 'Graduate', 'Postgraduate', 'Graduate'],
    'Age': [22, 27, 26, 23, 24, 30, 21]
}

# Create a DataFrame from the dictionary
emp_df = pd.DataFrame(data)

# Calculate the average age for graduate and postgraduate employees
average_age_graduate = emp_df[emp_df['Education'] == 'Graduate']['Age'].mean()
average_age_postgraduate = emp_df[emp_df['Education'] == 'Postgraduate']['Age'].mean()

# Display the results
print("Average Age of Graduate Employees:", average_age_graduate)
print("Average Age of Postgraduate Employees:", average_age_postgraduate)
```

```python
#TASK 8
#ADDITIONAL : DO A TENURE ANALYSIS OF ABOVE DATA
import pandas as pd
import numpy as np

# Create a DataFrame with hypothetical tenure data
data = {
    'Name': ['ABC', 'DEF', 'GHI', 'JKL', 'MNO', 'PQR', 'STU'],
    'Education': ['Graduate', 'Postgraduate', 'Postgraduate', 'Graduate', 'Graduate', 'Postgraduate', 'Graduate'],
    'Gender': ['Male', 'Female', 'Male', 'Other', 'Female', 'Female', 'Male'],
    'Tenure_Years': [3, 5, 2, 1, 4, 6, 2]  # Hypothetical tenure values in years
}

emp_df = pd.DataFrame(data)

# Calculate average tenure
average_tenure = emp_df['Tenure_Years'].mean()
print(f"Average Tenure: {average_tenure:.2f} years")

# Median tenure
median_tenure = emp_df['Tenure_Years'].median()
print(f"Median Tenure: {median_tenure:.2f} years")

# Tenure distribution
tenure_distribution = emp_df['Tenure_Years'].value_counts().sort_index()
print("\nTenure Distribution:")
print(tenure_distribution)

# Turnover rate (assuming 1-year period)
total_employees_at_start = len(emp_df)
employees_left = emp_df[emp_df['Tenure_Years'] < 1]
turnover_rate = len(employees_left) / total_employees_at_start * 100
print(f"\nTurnover Rate (within 1 year): {turnover_rate:.2f}%")

```

```python
#TASK 9
#9.1. From the following Cross-Sectional Dataframe 'df_cross_section': [05 marks]
#Company	2023	2024
#ZYX	123	321
#WVU	456	654
#TSR	789	987
#Create the following Panel Dataframe 'df_panel':
#Company	Year	Profit
#TSR	2023	789
#TSR	2024	987
#WVU	2023	456
#WVU	2024	654
#ZYX	2023	123
#ZYX	2024	321
import pandas as pd

# Given cross-sectional dataframe
data = {
    'Company': ['ZYX', 'WVU', 'TSR'],
    '2023': [123, 456, 789],
    '2024': [321, 654, 987]
}

df_cross_section = pd.DataFrame(data)

# Reshape the dataframe using melt
df_panel = pd.melt(df_cross_section, id_vars='Company', var_name='Year', value_name='Profit')

# Sort the dataframe by Company and Year
df_panel = df_panel.sort_values(by=['Company', 'Year'])

# Display the resulting panel dataframe
print(df_panel)


```

```python
#TASK 9
#ADDITIONAL : SORT ACCORDING TO INDEX NUMBER
# Sort the dataframe by index number
df_panel_sorted = df_panel.sort_index()

# Display the sorted dataframe
print(df_panel_sorted)
```
